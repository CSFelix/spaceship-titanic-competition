<h1 id="metric">üìè Metric</h1>

<br/>

This competition applies `Classification Accuracy` as the main metric to evaluate the results.

As far as the passengers have two possible outcomes to have been transported to another dimension, being `true` and `false`, this competition is a `Binary Classification Problem`.

In Binary Classification Problems, our models can have four predictions:

<br />

> **True Positive (TP)** - the model predicted `true`, and the real outcome is `true`; ‚úîÔ∏è

> **True Negative (TN)** - the model predicted `false`, and the real outcome is `false`; ‚úîÔ∏è

> **False Positive (FP)** - the model predicted `true`, and the real outcome is `false`; ‚ùå

> **False Negative (FN)** - the model predicted `false`, and the real outcome is `true`; ‚ùå

<br />

With this in mind, Classification Accuracy is calculated adding the True Positives and True Negatives, anbd dividing the result by the sum between True Positives, True Negatives, False Positives and False Negatives. Which means: 

$(TP + TN) / (TP + TN + FP + FN)$

So, consider that the real dataset has `150 True outcomes` and `150 False outcomes` and that my model has predicted `100 True Positives`, `100 True Negatives`, `50 False Positives` and `50 False Negatives`, the model's accuracy will be:

$(TP + TN) / (TP + TN + FP + FN)$

$(100 + 100) / (100 + 100 + 50 + 50)$

$200 / 300 == 2 / 3 =~ 0.67 == 67$%

<h1 id='treating-outliers'>üõë Treating Outliers</h1>

For Outliers Detection, let's use `K-Nearest Neighbors (KNN)` Algorithm from PyOD library. This Algorithm works calculating the distance between each element into the dataset and, the points that are closer, are classified into the same class. So, all elements that have significant high values and are closer to each other, are considered as outliers.

> **OBS.:** Unfortunately, this library is not native on Kaggle, so we have to install it on our device using the `!` operator.

````python
!pip install pyod
````

````python
from pyod.models.knn import KNN

# Defining outliers contamination fraction and the Classifier Model
# PyOD KNN Documentation: https://pyod.readthedocs.io/en/latest/_modules/pyod/models/knn.html
outliers_fraction = 0.10
classifier_name = 'KNN'
classifier = {
    classifier_name: KNN(
                        contamination=outliers_fraction
                        , method='largest'
                        , n_neighbors=5
                        , n_jobs=4
                    )
}
````

````python
# Training KNN Model and making Predictions
classifier[classifier_name].fit(X_train)

# Getting the Binary Labels (0: inliers, 1: outliers)
X_train_predictions = classifier[classifier_name].labels_
X_valid_predictions = classifier[classifier_name].predict(X_valid)
````

````python
# Getting the Outliers Row Indexes
outliers_indexes_X_train = [
    index for index in range(0, len(X_train_predictions))
    if X_train_predictions[index] == 1
]

outliers_indexes_X_valid = [
    index for index in range(0, len(X_valid_predictions))
    if X_valid_predictions[index] == 1
]
````

````python
# Getting the Outliers Dataset
outliers_rows_X_train = X_train.iloc[outliers_indexes_X_train]
outliers_rows_X_valid = X_valid.iloc[outliers_indexes_X_valid]

percentage_outliers_X_train = len(outliers_indexes_X_train) * 100 / len(X_train)
percentage_outliers_X_valid = len(outliers_indexes_X_valid) * 100 / len(X_valid)

print('Outliers Percentage - Train Dataset:', percentage_outliers_X_train)
print('Outliers Percentage - Validation Dataset:', percentage_outliers_X_valid)
````

Since less than 15% of the datas are outliers, there is no problem to just drop them from our train and validation datasets.

````python
# Dropping outlier rows (X - Train and Validation Datasets)
X_train.drop(X_train.index[outliers_indexes_X_train], inplace=True)
X_valid.drop(X_valid.index[outliers_indexes_X_valid], inplace=True)

# Dropping outlier rows (y - Train and Validation Datasets)
y_train.drop(y_train.index[outliers_indexes_X_train], inplace=True)
y_valid.drop(y_valid.index[outliers_indexes_X_valid], inplace=True)
````

<b>Mutual Information (MI)</b>

First, we are going to check out which features are most important to pass to the model.

````python
from sklearn.feature_selection import mutual_info_regression

# Function to Plot MI Scores
def plot_mi_scores(scores):
    """
    Plots Mutual Information Scores in Ascending Order
    """
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")
    plt.show()
    
# Convert "int8" features to int64" in train dataset
X_train['HomePlanet'] = X_train['HomePlanet'].astype(int)
X_train['Destination'] = X_train['Destination'].astype(int)

# Convert "int8" features to int64" in validation dataset
X_valid['HomePlanet'] = X_valid['HomePlanet'].astype(int)
X_valid['Destination'] = X_valid['Destination'].astype(int)

# Convert "int8" features to int64" in test dataset
full_test['HomePlanet'] = full_test['HomePlanet'].astype(int)
full_test['Destination'] = full_test['Destination'].astype(int)
````

````python
# Getting the discrete (integer) and continuous (float) features
discrete_features = [
    feature for feature in X_train.columns
    if X_train[feature].dtype == 'int64'
]

continuous_features = [
    feature for feature in X_train.columns
    if X_train[feature].dtype == 'float64'
]

# Creating copy datasets for discrete and continuous features
discrete_features_list = X_train[discrete_features].dtypes == 'int64'
continuous_features_list = X_train[continuous_features].dtypes == 'float64'
````

````python
# Calculating Features MI Scores
mi_scores_discrete_features = mutual_info_regression(X_train[discrete_features], y_train, discrete_features=discrete_features_list, random_state=SEED)
mi_scores_discrete_features = pd.Series(mi_scores_discrete_features, name='MI Scores - Discrete Features', index=discrete_features)
mi_scores_discrete_features = mi_scores_discrete_features.sort_values(ascending=False)

mi_scores_continuous_features = mutual_info_regression(X_train[continuous_features], y_train, discrete_features=continuous_features_list, random_state=SEED)
mi_scores_continuous_features = pd.Series(mi_scores_continuous_features, name='MI Scores - Continuous Features', index=continuous_features)
mi_scores_continuous_features = mi_scores_continuous_features.sort_values(ascending=False)

# Plotting the Results
plt.figure(dpi=100, figsize=(15, 20))
plot_mi_scores(mi_scores_discrete_features)

plt.figure(dpi=100, figsize=(10, 10))
plot_mi_scores(mi_scores_continuous_features)
````

It looks like only `VIP` feature has a significant low MI value (less than `0.02`), so let's drop this feature from our datasets.

````python
# Dropping 'VIP' feature from the datasets
X_train.drop('VIP', axis=1, inplace=True)
X_valid.drop('VIP', axis=1, inplace=True)
full_test.drop('VIP', axis=1, inplace=True)
````

<b>K-Means Clustering</b>

Now, we are going to classify the datasets' elements into groups!!

````python
from sklearn.cluster import KMeans

# First, let's find out what's the perfect number of clusters applying
# the Elbow and WCSS Methods
#
# Check out for more information about Elbow and WCSS Methods here:
# https://www.geeksforgeeks.org/elbow-method-for-optimal-value-of-k-in-kmeans/
wcss = []

for i in range(1,7):
    kmeans = KMeans(
        n_clusters=i
        , init='k-means++'
        , max_iter=300
        , n_init=10
        , random_state=SEED
    )
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
    
# Plotting the results
plt.plot(range(1,7), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
````

Since the plot has an elbow with `2 clusters` only, we are going to use 2 clusters to classify our datasets.

````python
kmeans = KMeans(
    n_clusters=2
    , init='k-means++'
    , max_iter=300
    , n_init=10
    , random_state=SEED
)

# Creating the 'Cluster' feature
X_train.insert(loc=0, column='Cluster', value=kmeans.fit_predict(X_train))
X_valid.insert(loc=0, column='Cluster', value=kmeans.fit_predict(X_valid))
full_test.insert(loc=0, column='Cluster', value=kmeans.fit_predict(full_test))

# Peeking the first five rows from train training
X_train.head()
````

<h1 id='baseline-ml-models'>ü§ñ Baseline ML Models</h1>

Let's create two baseline models:

```
- XGBoost Classifier
- Deep Learning Model
```

But before everything, we have to prepare a pipeline to `Standard/Scale/Normalize` the datas and create a function to evaluate the models.

----

<b>Pipelines</b>

````python
# Pipelines
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.compose import ColumnTransformer

# General Transformer
transformer = Pipeline(steps=[
    #('scaler', MinMaxScaler())
    #('scaler', RobustScaler())
    ('scaler', StandardScaler())
    #('scaler', Normalizer())
])

# General Preprocessor
preprocessor = ColumnTransformer(transformers=[('numerical', transformer, X_train.columns)])
````

----

<b>Evalutation Function</b>

````python
from sklearn.metrics import accuracy_score
accuracy = lambda real_values, predictions: accuracy_score(real_values, predictions)
````

----

<b>Baseline Model 1: XGBoost Classifier</b>

````python
import xgboost as xgb

# Preprocessing the datas
xgb_X_train = preprocessor.fit_transform(X_train)
xgb_X_valid = preprocessor.transform(X_valid)
xgb_full_test = preprocessor.transform(full_test)

# Creating the Model
xgb_model = xgb.XGBClassifier(
     objective='binary:logistic'
    , n_estimators=250
    , learning_rate=0.10
    , colsample_bytree=0.70
    , max_depth=3
    , n_jobs=4
)
````

````python
# Training the Model and Making Predictions with Validation dataset
xgb_model.fit(
    xgb_X_train, y_train
    , early_stopping_rounds=5
    , eval_set=[(xgb_X_valid, y_valid)]
    , verbose=False
)

print('Training Done!')

xgb_predictions = xgb_model.predict(xgb_X_valid)

print('Predictions Done!')
````

````python
# Accuracy
xgb_accuracy = accuracy(y_valid, xgb_predictions)
print('Accuracy: ', round(xgb_accuracy * 100, 2), '%', sep='')
````

````python
# Train and Validation Scores
print('Train Score: %.2f%%' % (xgb_model.score(xgb_X_train, y_train) * 100))
print('Validation Score: %.2f%%' % (xgb_model.score(xgb_X_valid, y_valid) * 100))
````

----

<b>Baseline 2: Deep Learning Model</b>

````python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping

# Creating the Model
dl_model = keras.Sequential([
    # hidden layers
    layers.BatchNormalization(input_shape=[len(X_train.columns)]),
    layers.Dense(units=4, activation='relu'),
    
    layers.BatchNormalization(),
    layers.Dense(units=4, activation='relu'),
    
    layers.BatchNormalization(),
    layers.Dense(units=4, activation='relu'),

    # output layer
    layers.Dense(units=1, activation='sigmoid'),
])

# Defining the Optimizer and Loss Functions
dl_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['binary_accuracy'],
)

# Early Stopping
early_stopping = EarlyStopping(
    min_delta=0.001,
    patience=10,
    restore_best_weights=True,
)
````

````python
# Fitting the Model and Making Predictions
history = dl_model.fit(
    X_train, y_train,
    validation_data=(X_valid, y_valid),
    batch_size=512,
    epochs=1000,
    callbacks=[early_stopping],
    verbose=False
)

print('Training Done!')

dl_predictions = dl_model.predict(X_valid)

print('Predictions Done!')
````

````python
# Accuracy
history_df = pd.DataFrame(history.history)
history_df.loc[:, ['loss', 'val_loss']].plot(title='Cross-Entropy');
history_df.loc[:, ['binary_accuracy', 'val_binary_accuracy']].plot(title='Accuracy');
````

Since both models got a similar accuracy (around`78%`), let's keep going with both of them. But, over `Machine Learning Explainability` step, we will considedr the XGBClassifier only.
